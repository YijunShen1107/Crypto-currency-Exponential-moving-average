```python
import numpy as np 
import pandas as pd
import datetime
import yfinance as yf
from sklearn.linear_model import LinearRegression
import warnings
warnings.filterwarnings('ignore')
```

```python
def collect(name, period):
    data = yf.download(name, period = period, interval = '1d') 
    data.rename(columns={'Adj Close': name}, inplace=True)
    data = data[name].dropna()
    return data 


#names = ["BTC-USD","XRP-USD","ETH-USD","BCH-USD","ADA-USD","XEM-USD","LTC-USD","TRX-USD","XLM-USD","MIOTA-USD","DASH-USD","EOS-USD",
#"XMR-USD","NEO-USD","QTUM-USD","BTG-USD","SC-USD","BCN-USD","ZEC-USD","XVG-USD","OMG-USD","BTS-USD","PPT-USD","DOGE-USD","DCN-USD","BNB-USD","SNT-USD","LINK-USD","USDC-USD","VET-USD"]
#names = ["BTC-USD","XRP-USD","ETH-USD","ADA-USD","XEM-USD","LTC-USD","TRX-USD","XLM-USD","MIOTA-USD","DASH-USD",
#"XMR-USD","NEO-USD","SC-USD","BCN-USD","ZEC-USD","XVG-USD","OMG-USD","BTS-USD","PPT-USD","DOGE-USD","DCN-USD","BNB-USD","SNT-USD","LINK-USD","USDC-USD","VET-USD"]
names = ["BTC-USD","XRP-USD","ETH-USD","ADA-USD","XEM-USD","LTC-USD","TRX-USD","XLM-USD","MIOTA-USD","DASH-USD",
"XMR-USD","NEO-USD","SC-USD","BCN-USD","ZEC-USD","XVG-USD","OMG-USD","BTS-USD","DCN-USD","BNB-USD","SNT-USD","LINK-USD","VET-USD","HT-USD","BCH-USD","FIL-USD","EOS-USD","THETA-USD",]
all_data = []


for name in names:
    data = collect(name, period='4y')
    all_data.append(data)    

final_table = pd.concat(all_data, axis=1).reindex(data.index)
final_table = final_table.iloc[:1000]

print(final_table)
#final_table.to_csv("pricedata.csv")
```

```python
#df = final_table
# Create 20 days exponential moving average column
def signalgeneration(data):
    df = pd.DataFrame(index = data.index)
    for name in names:
        df[name + '20_EMA'] = data[name].ewm(span = 20, adjust = False).mean()
        df[name + '50_EMA'] = data[name].ewm(span = 50, adjust = False).mean()  
        df[name + 'Signal'] = 0.0
        df[name + 'Signal'] = np.where(df[name +'20_EMA'] > df[name +'50_EMA'], 1.0, -1.0)
    sig = []
    for name in names:
        data = df[name + 'Signal']
        sig.append(data)
    signal = pd.concat(sig, axis=1).astype(np.float32)
    return signal
#get a table with only signals
   # df[name + 'Position'] = df[name + 'Signal'].diff()
#if the position is -1, meaning we obtain the shorting signal and we need to switch from long to short on the next day
#if the position is 1, meaning we obtain the longing signal and we need to switch from short to long on the next day
```

```python
position = signalgeneration(final_table)
```

```python
#refinement 2: daily select only two best performed crypto
def SW_signalgenerator(df, ma1, ma2):
    #ma2 is the ma with more days
    #for the input ma1, ma2, we just need to plug in number of days we need for each moving average
    ma1 = df.rolling(ma1).mean().dropna(axis = 0, how ='all')
    ma2 = df.rolling(ma2).mean().dropna(axis = 0, how = 'all')
    signal = (np.log(ma2)-np.log(ma1)).dropna(axis = 0, how = 'all')
    #normalize signal
    norm_sign = 0.02*((signal.T - signal.mean(axis = 1))/signal.std(axis=1)).T
    return norm_sign

position = SW_signalgenerator(final_table, 20, 50)

def backtest1(df, sign):
    r = df.pct_change()

    b = pd.DataFrame(columns=['short crypto','long crypto', 
                              'short position', 'long position',
                              'short change','long change','return', 'net worth'],index=sign.index)
    for idx in sign.index:
        # sale 
        b['short crypto'].loc[idx] = sign.loc[idx][sign.loc[idx]==sign.loc[idx].max()].index[0]
        b['short position'].loc[idx] = sign.loc[idx].max()
        b['short change'].loc[idx] = r[b['short crypto'].loc[idx]].loc[idx]
        # buy
        b['long crypto'].loc[idx] = sign.loc[idx][sign.loc[idx]==sign.loc[idx].min()].index[0]
        b['long position'].loc[idx] = sign.loc[idx].min()
        b['long change'].loc[idx] = r[b['long crypto'].loc[idx]].loc[idx]
        # daily return
        b['return'] = b['short change']*b['short position'] - b['long position']*b['long change']
    b['net worth'] = (b['return']+1).cumprod()
    
    return b
```

```python
def maxdrawdown(return_list):
    """maxdrawdown"""
    maxac=np.zeros(len(return_list))
    b=return_list[0]
    for i in range(0,len(return_list)): #Goes over the list, when larger than the previous value, set b
        if return_list[i]>b:
            b=return_list[i]
        maxac[i]=b
    #print(maxac)
    i=np.argmax((maxac-return_list)/maxac) #stop position
    if i == 0:
        return 0
    j = np.argmax(return_list[:i])  #start position
    maxdd = (return_list[j] - return_list[i]) / (return_list[j])
    return maxdd
maxdrawdown(portfolio['networth'])#0.9428505873117996
```
